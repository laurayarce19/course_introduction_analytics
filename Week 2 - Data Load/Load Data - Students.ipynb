**JSON**

Los tipos de datos disponibles con JSON son:

Números: Se permiten negativos y pueden contener parte fraccional separada por puntos. Ejemplo: 123.456

Cadenas: Representan secuencias de cero o más caracteres. Se ponen entre doble comilla y se permiten cadenas de escape. Ejemplo: "Hola"

Booleanos: Representan valores booleanos y pueden tener dos valores: true y false

null: Representan el valor nulo.

Array: Representa una lista ordenada de cero o más valores los cuales pueden ser de cualquier tipo. Los valores se separan por comas y el vector se mete entre corchetes. Ejemplo ["juan","pedro","jacinto"]

Objetos: Son colecciones no ordenadas de pares de la forma <nombre>:<valor> separados por comas y puestas entre llaves. El nombre tiene que ser una cadena entre comillas dobles. El valor puede ser de cualquier tipo.

**DESDE BIGQUERY**

**PASOS:**

Cuenta de servicio
- Ir en el menu principal a **API y servicios**
- Dar click en la parte superior **CREAR CREDENCIALES**
- Seleccionar **cuenta de servicio** (service account)
> - Asignar un nombre: es libre
> - Descripción de la cuenta de servicio: es libre
- Dar click en **CREAR Y CONTINUAR**
- Seleccionar el rol que va tener la cuenta, el rol son conjuntos de permisos, ejemplo: Básico-Propietario
- El tercer paso dejarlo vacio
- Dar click en **LISTO**

- Dar click en el **correo de la cuenta de servicio** creada
- En el menu de arriba dar click en **CLAVES**
- Dar click en **AGREGAR CLAVE**
- Dar click en **Crear clave nueva**
- Seleccionar JSON y darle **crear**

```python
from google.cloud import bigquery 
from google.oauth2 import service_account
```

```python
credentials_path = 'escribir ruta de las credenciales'   # direccion local de archivo JSON con las credenciales del proyecto        
credentials = service_account.Credentials.from_service_account_file(credentials_path) # asignar las credenciales a la service account
```

```python
project_id = 'escribir ruta del id del proyecto'
client = bigquery.Client(credentials= credentials, project=project_id) # creacion del client para trabajar en python
```

```python
consulta = '''SELECT sexo, sum(cantidad) as total FROM `curso4-379521.hurto.hurto_carro`
group by sexo '''
```

```python
df = client.query(consulta).to_dataframe()
df.head()
```

**Otros formatos**

```python
# formato pickel: es un formato binario, permite comprimir la información, es muy utilcuando se tienen tablas grandes
df.to_pickle('/content/hurtos.pkl') # guarde la base en .pkl
df = pd.read_pickle('/content/hurtos.pkl')
df
```

```python
# Formato parquet: puede usarse en ambientes de Big Data
df.to_parquet('/content/hurtos.parquet')
df = pd.read_parquet('/content/hurtos.parquet')
df
```

```python
# Formato HDFS: puede usarse en ambientes de Big Data con Hadoop
df.to_hdf('/content/hurtos.h5', key = 'agua', format = 'table')
df = pd.read_hdf('/content/hurtos.h5')
df
```
