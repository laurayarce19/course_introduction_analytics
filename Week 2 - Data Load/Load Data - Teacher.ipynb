**JSON**

Los tipos de datos disponibles con JSON son:

Números: Se permiten negativos y pueden contener parte fraccional separada por puntos. Ejemplo: 123.456

Cadenas: Representan secuencias de cero o más caracteres. Se ponen entre doble comilla y se permiten cadenas de escape. Ejemplo: "Hola"

Booleanos: Representan valores booleanos y pueden tener dos valores: true y false

null: Representan el valor nulo.

Array: Representa una lista ordenada de cero o más valores los cuales pueden ser de cualquier tipo. Los valores se separan por comas y el vector se mete entre corchetes. Ejemplo ["juan","pedro","jacinto"]

Objetos: Son colecciones no ordenadas de pares de la forma <nombre>:<valor> separados por comas y puestas entre llaves. El nombre tiene que ser una cadena entre comillas dobles. El valor puede ser de cualquier tipo.

# **DESCARGA DE DATOS**

**Directo al ordenador**

```python
acc.to_csv('/content/drive/MyDrive/accidentes.csv', index = False) # Guardamos el archivo como accidentes.csv
files.download('/content/drive/My Drive/accidentes.csv') # Descargamos el archivo en nuestra maquina local
```

**Otros formatos**

```python
# Formato pickel: es un formato binario, permite comprimir la información, por lo cual, es util cuando se tienen tablas grandes
acc.to_pickle('/content/drive/My Drive/accidentes.pkl')
acc = pd.read_pickle('/content/drive/My Drive/accidentes.pkl')
acc.head(3)
```

```python
# Formato parquet: puede usarse en ambientes de Big Data cuando las bases pertenecen a Apache
acc.to_parquet('/content/drive/My Drive/accidentes.parquet')
acc = pd.read_parquet('/content/drive/My Drive/accidentes.parquet')
acc.head(3)
```

```python
# Formato HDFS: puede usarse en ambientes de Big Data con Hadoop
acc.to_hdf('/content/drive/My Drive/accidentes.h5', key = 'acc', format= 'table')
acc = pd.read_hdf('/content/drive/My Drive/accidentes.h5')
acc.head(3)
```
