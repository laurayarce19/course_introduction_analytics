### CONTROL DEL FLUJO

**Flujo:** Es el orden en que se ejecutan las instrucciones o líneas de código. En Python, las líneas de código se ejecutan de arriba hacia abajo. Si no hay nada que interrumpa o modifique el orden de las instrucciones, la ejecución de estas será línea por línea consecutivamente hasta la última línea de código.

**Indentación:** La indentación se puede definir como el espacio en blanco que se sitúa al inicio de una línea de código. La cantidad de espacios a los que corresponda la indentación está definida por la convención usada en el lenguaje de programación. En Python suele usarse una tabulación con la tecla TAB. La cantidad de indentaciones que tenga una línea se conoce como nivel de indentación.

### SENTENCIAS MÁS UTILIZADAS

**CONDICIONAL: IF**

Para definir un if, primero se debe sentenciar usando la palabra clave if; luego, entre paréntesis “()” o no, se indica la condición que permitirá la ejecución del bloque de código (una condición suele ser una operación relacional y/o lógica que debe dar como resultado True para ejecutar la sentencia) y por último se agregan dos puntos “:”. Tras esto, se hace un salto de línea a partir del cual se escribirá un bloque de código con UNO de indentación más, el cual será el bloque de código a ejecutar bajo el cumplimiento de la condición.

```python
# Definir variables
peso = int(input("Cuánto pesas (kl): "))
estatura = float(input("Cuál es tu estatura (m): "))

# Calcular índice de masa corporal
imc = peso/(estatura**2)
if imc < 18.5:
  print('Peso inferior al normal')
elif 18.5 <= imc < 25:
  print('Peso normal')
elif 25 <= imc < 30:
  print('Peso superior al normal')
else:
  print('Obesidad')
```

**BUCLE: WHILE**

Se usa para ejecutar un bloque de código en repetidas veces si una condición se cumple. Para definir un while primero se escribe la palabra clave while; luego, entre paréntesis “()”, se indica la condición (la cual permitirá iniciar el bucle) y, por último, se agregan dos puntos “:”. Después de esto se escribe el bloque de código con su respectiva indentación. En cada iteración de un bucle se ejecuta su bloque de código siempre y cuando la condición se cumpla cada que se evalúe.

```python
lim_turn = int(input('Ingrese el número de turnos'))
actual_turn = 1
while(actual_turn <= lim_turn):
   print("Turno " + str(actual_turn)) 
   actual_turn += 1
print("No hay más servicio por hoy")
```

**BREAK**

La sentencia break, como lo indica su nombre (romper, en inglés), se utiliza para romper un bucle o ciclo en una línea de código específica. Cuando se habla de romper un bucle, se refiere a dar por terminado su flujo, es decir, omitir el resto de las líneas de su bloque de código y el resto de las iteraciones para continuar en la línea que sigue fuera del ciclo. Un break se define en una sola línea de código con la palabra clave break.

```python
num_resources = 
while(num_resources > 0):
  print("Utilizar recurso")
  num_resources -= 1
  if(num_resources == 4): 
    break
  print("Recursos disponibles: " + str(num_resources))
```

**CONTINUE**

La sentencia continue (continuar, en inglés) se utiliza para terminar la repetición que se esté realizando en el momento, es decir, cuando se usa un continue se va a omitir el resto de las líneas de código que estén dentro del ciclo y va a regresar al inicio de este, donde se evalúa nuevamente la condición que permite entrar en el ciclo

**SETS**

Los sets son la estructura de datos que representan los conjuntos matemáticos. Los sets presentan varias características particulares en comparación con las estructuras de datos vistas: son desordenadas, es decir, no se mantiene registro alguno del orden en el que se ingresan los datos, entonces no existe un subíndice el cual se asigna a cada dato. Los datos de un set son únicos, es decir, no se pueden tener múltiples datos iguales. Finalmente, los sets permiten realizar operaciones de conjuntos entre ellos.



```python
mi_set = {}
mi_set
```

```python
# Métodos de los sets: Unión  
set_1 = {}
set_2 = {}
union_set = set_1.union(set_2)
union_set # devolverá un set que contiene TODOS los datos de ambos sets.
```

```python
# Métodos de los sets: Intersección
set_1 = {}
set_2 = {}
intersection_set = set_1.intersection(set_2)
intersection_set # devolverá un set que contiene los datos que pertenecen a ambos sets A LA MISMA VEZ.
```

```python
# Métodos de los sets: Diferencia
set_1 = {}
set_2 = {}
difference_set = set_1.difference(set_2)
difference_set # devuelve un set con los datos que pertenecen al primer set pero NO pertenecen al segundo set. 
```

```python
# Métodos de los sets: Diferencia simétrica
set_1 = {}
set_2 = {}
symmetric_difference_set = set_1.symmetric_difference(set_2)
symmetric_difference_set # devuelve como resultado TODOS los datos de ambos sets, pero excluyendo los datos que hacen parte de ambos sets A LA MISMA VEZ. 
```

### LIBRERÍAS Y MÓDULOS

**Librerías:** Las librerías son un conjunto de soluciones predefinidas. También puede decirse que son una suite de herramientas que otros han desarrollado para ser utilizadas en cualquier nuevo proyecto para agilizar algunas características o funcionalidades de una aplicación o sistema en construcción.

**Tipos de librerías:** Según la documentación oficial de Python, hay dos tipos de librerías. Las librerías internas, denominadas built-in functions (o librerías del sistema), en las que no es necesario realizar procesos de instalación para su uso, y librerías externas, en las que es obligatorio llevar a cabo procesos de instalación adicional para acceder a ellas.

**Sintaxis**: se invoca dicha librería a través de la sentencia (import)
en minúscula. Es importante resaltar que dichas librerías tienen,
por lo general, elementos internos tales como métodos y
funciones, en los cuales, para usar aquellas implementaciones, se
debe invocar mediante la nomenclatura del punto (.).

**Modulo:** Una ventaja de utilizar las sentencias from e import es
específicamente poder limitar el coste computacional de
módulos, métodos y clases innecesarias en proyectos
complejos, dado que utilizar todas las clases o métodos de una
librería podría traer consigo una gran carga computacional y
hacer que sea lenta en su rendimiento.
